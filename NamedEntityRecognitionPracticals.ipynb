# Named Entity Recognition

Examples
1. Person
2. Place or Location
3. Date
4. Time
5. Money
6. Organization
7. Percent

```python
sentence = """The eiffel Tower was built from 1887 to 1889
by French Engineer Gustave Eiffel,
whose company specialized in building
metal frameworks and structures."""
print(sentence)
```

```python
import nltk
nltk.download('punkt_tab')
words = nltk.word_tokenize(sentence)
```

```python
nltk.download('averaged_perceptron_tagger_eng')
tag_elements=nltk.pos_tag(words)
print(tag_elements)
```

```python
nltk.download('maxent_ne_chunker_tab')
nltk.download('words')
print(nltk.ne_chunk(tag_elements))
nltk.ne_chunk(tag_elements).draw()
```
